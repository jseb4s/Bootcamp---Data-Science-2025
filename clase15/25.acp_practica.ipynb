```python
# Análisis de Componentes Principales (ACP) con dataset de vidrios

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import seaborn as sns

df = pd.read_csv("glass.csv")
```

```python
df.head()
```

```python
X = df.drop(columns=['Type'])
```

```python
X
```

```python
# Estandarizacion
scaler = StandardScaler() # Objeto de estandarizacion. Puede variar si queremos minmax, etc.
X_scaled = scaler.fit_transform(X) # aplicar estandarizacion en nuestros datos
```

```python
X_scaled
```

```python
# ACP

"""
    PCA(n_components=2)  # Conserva solo los 2 primeros componentes

    PCA(n_components=0.95)  # Conserva el número mínimo de componentes necesarios para explicar el 95% de la varianza

    'mle' → máximo número basado en información de verosimilitud (requiere svd_solver='full')
    PCA(n_components='mle', svd_solver='full')

    None (por defecto): conserva todos los componentes posibles
    PCA(n_components=None)
    
    PCA(svd_solver='auto')      # (por defecto)
    PCA(svd_solver='full')      # para datasets pequeños
    PCA(svd_solver='randomized')  # para datasets grandes
    PCA(svd_solver='arpack')    # para cálculo basado en eigenvalores
    
    
"""

pca = PCA(n_components = 1)

X_pca = pca.fit_transform(X_scaled)
```

```python
X_scaled
```

```python
X_pca
```

```python
# Varianza explicada
explained_var = pca.explained_variance_ratio_
cumulative_var = np.cumsum(explained_var)
cumulative_var
```

```python
plt.figure(figsize=(8, 5))
plt.bar(range(1, len(explained_var) + 1), explained_var, alpha=0.7, align='center', label='Varianza explicada')
plt.step(range(1, len(cumulative_var) + 1), cumulative_var, where='mid', label='Varianza acumulada', linestyle='--')
plt.xlabel('Numero de Componentes Principales')
plt.ylabel('Proporcion de Varianza Explicada')
plt.title('Varianza Explicada por Componente (ACP)')
plt.legend(loc='best')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# cos² (calidad de representación) para individuos en PC1 y PC2

# distancia total al origen para cada individuo
distancia_total = np.sum(X_pca**2, axis=1)

# cos² para cada componente (solo PC1 y PC2)
cos2_ind = (X_pca[:, :2]**2).T / distancia_total

cos2_df = pd.DataFrame(cos2_ind.T, columns=['cos2_PC1', 'cos2_PC2'], index=df.index)
cos2_df['cos2_total'] = cos2_df['cos2_PC1'] + cos2_df['cos2_PC2']
cos2_df
```

¿Qué tan bien está representado un individuo (observación) en un componente?

Su valor va de 0 a 1 (o de 0% a 100%).

- Un cos² cercano a 1 → el individuo está muy bien proyectado en ese eje.

- Un cos² cercano a 0 → está mal representado (queda fuera del plano, como si fuera una sombra borrosa).

```python
# DataFrame con componentes
df_pca = pd.DataFrame(X_pca, columns=[f'PC{i+1}' for i in range(X_pca.shape[1])])
df_pca['Type'] = df['Type']
df_pca
```

```python
df_plot = df.copy()
df_plot['PC1'] = df_pca['PC1']
df_plot['PC2'] = df_pca['PC2']
df_plot['PC3'] = df_pca['PC3']
df_plot['PC4'] = df_pca['PC4']
```

```python
df_plot
```

```python
# Cargas (autovectores)
loadings = pd.DataFrame(pca.components_.T,
                        columns=[f'PC{i+1}' for i in range(pca.n_components_)],
                        index=X.columns)
loadings
```

Esta línea crea un DataFrame que muestra cuánto contribuye cada variable original a cada componente principal. En términos simples:

"¿Qué tanto participa cada variable original en la construcción de cada PC?".

Los valores se ubicaran entre -1 y 1 aprox. por la estandarización realizada.

Los valores son coeficientes de la combinación lineal que define cada componente.

---

```python
# Circulo de correlaciones (PC1 y PC2)
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(8, 8))

# vectores para cada variable
for i, var in enumerate(loadings.index):
    plt.arrow(0, 0, loadings.iloc[i, 0], loadings.iloc[i, 1],
              head_width=0.02, color='blue', alpha=0.7)
    plt.text(loadings.iloc[i, 0]*1.15, loadings.iloc[i, 1]*1.15, var,
             ha='center', va='center', fontsize=9)

# lineas centrales
plt.axhline(0, color='gray', linestyle='--')
plt.axvline(0, color='gray', linestyle='--')

# circulo unitario
circle = plt.Circle((0, 0), 1, color='gray', fill=False, linestyle='--', linewidth=1)
plt.gca().add_patch(circle)
plt.title('Círculo de Correlaciones (PC1 vs PC2)')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.grid(True)
plt.xlim(-1.1, 1.1)
plt.ylim(-1.1, 1.1)
plt.gca().set_aspect('equal', adjustable='box')
plt.tight_layout()
plt.show()
```

Es un gráfico en el plano PC1 vs PC2 donde se proyectan las variables originales estandarizadas como vectores (flechas). Este gráfico muestra:

- Dirección: indica con qué componente principal se asocia cada variable.

- Longitud: indica qué tan bien está representada la variable en el plano (cuánto aporta).

- Ángulo entre flechas: indica el tipo de relación entre las variables originales.

- “Cada flecha representa una variable. Si la flecha llega cerca del borde del círculo, significa que las dos primeras componentes explican muy bien esa variable. Si se queda corta, esa variable está mejor representada en otra dimensión.”

# ACM

Imagina que estás en un mapa de dos dimensiones. Cada categoría (como “Joven” o “Producto B”) es una ciudad. El ACM te dice dónde queda cada ciudad en ese mapa. Las ciudades cercanas representan categorías que tienden a aparecer juntas en los datos reales.

```python
import pandas as pd

# Crear un pequeño dataset de consumidores
data_acm = {
    'Género': [
        'Mujer', 'Hombre', 'Mujer', 'Hombre', 'Mujer', 'Hombre', 'Mujer', 'Hombre',
        'Mujer', 'Hombre', 'Mujer', 'Hombre', 'Mujer', 'Hombre', 'Mujer', 'Hombre',
        'Mujer', 'Hombre', 'Mujer', 'Hombre', 'Mujer', 'Hombre', 'Mujer', 'Hombre',
        'Mujer', 'Hombre', 'Mujer', 'Hombre', 'Mujer', 'Hombre'
    ],
    'Edad': [
        'Joven', 'Joven', 'Adulto', 'Adulto', 'Adulto', 'Mayor', 'Mayor', 'Joven',
        'Joven', 'Adulto', 'Mayor', 'Adulto', 'Joven', 'Mayor', 'Adulto', 'Mayor',
        'Mayor', 'Adulto', 'Joven', 'Joven', 'Adulto', 'Joven', 'Mayor', 'Adulto',
        'Adulto', 'Mayor', 'Joven', 'Adulto', 'Mayor', 'Joven'
    ],
    'Producto_Fav': [
        'A', 'B', 'A', 'C', 'C', 'B', 'C', 'A',
        'B', 'C', 'C', 'A', 'A', 'B', 'C', 'C',
        'A', 'B', 'B', 'C', 'A', 'B', 'C', 'C',
        'A', 'B', 'B', 'A', 'C', 'C'
    ],
    'Canal_Compra': [
        'Online', 'Online', 'Presencial', 'Presencial', 'Online', 'Presencial', 'Online', 'Presencial',
        'Online', 'Presencial', 'Presencial', 'Online', 'Presencial', 'Presencial', 'Online', 'Online',
        'Online', 'Presencial', 'Online', 'Presencial', 'Online', 'Presencial', 'Online', 'Presencial',
        'Online', 'Presencial', 'Presencial', 'Online', 'Online', 'Online'
    ],
    'Frecuencia_Compra': [
        'Alta', 'Media', 'Alta', 'Baja', 'Media', 'Baja', 'Alta', 'Media',
        'Alta', 'Alta', 'Media', 'Media', 'Baja', 'Alta', 'Alta', 'Media',
        'Baja', 'Alta', 'Alta', 'Media', 'Baja', 'Alta', 'Media', 'Media',
        'Baja', 'Media', 'Alta', 'Alta', 'Baja', 'Alta'
    ]
}

df = pd.DataFrame(data_acm)
df
```

```python
!pip install -q prince
import prince
```

```python
# Inicializar el modelo de ACM
mca = prince.MCA(n_components=2, random_state=42)
```

```python
# Ajustar el modelo
mca = mca.fit(df)
```

```python
mca
```

```python
# Transformar los datos a coordenadas factoriales
coords = mca.transform(df)
coords
```

Cada individuo es ahora representado por un punto en un espacio de baja dimensión (por ejemplo, 2 dimensiones).

El resultado coords es un nuevo DataFrame numérico, donde:

- Las filas representan los individuos.
- Las columnas son las dimensiones principales del ACM (Dimensión 1, Dimensión 2, ...).



```python
# Calculo de inercia

# Coordenadas de las columnas (categorias)
col_coords = mca.column_coordinates(df)

# Calcular inercia (varianza) por eje usando la suma de los cuadrados de las coordenadas de las categorías
inertias = (col_coords**2).sum(axis=0)

# Normalizar para obtener porcentaje de inercia explicada
inertia_ratio = inertias / inertias.sum()

for i, ratio in enumerate(inertia_ratio):
    print(f"Dimensión {i+1}: {ratio:.4f} ({ratio*100:.2f}%) de inercia explicada")
```

```python
import matplotlib.pyplot as plt

# Plano factorial

# Los individuos (las observaciones o filas del DataFrame original).
# Las categorías (valores posibles de las variables cualitativas).


# coordenadas de individuos y categorias
row_coords = mca.row_coordinates(df)
col_coords = mca.column_coordinates(df)

# plot
fig, ax = plt.subplots(figsize=(8, 6))

# individuos
ax.scatter(row_coords[0], row_coords[1], color='blue', label='Individuos', s=50)
# categorias
ax.scatter(col_coords[0], col_coords[1], color='red', label='Categorias', s=80)
# etiquetas de categorias
for i, txt in enumerate(col_coords.index):
    ax.text(col_coords.iloc[i, 0], col_coords.iloc[i, 1], txt, color='black', fontsize=9, ha='center')
# etiquetas de individuos
for i, txt in enumerate(row_coords.index):
    ax.text(row_coords.iloc[i, 0], row_coords.iloc[i, 1], txt, color='gray', fontsize=8)
    
# ejes
ax.axhline(0, color='gray', linestyle='--')
ax.axvline(0, color='gray', linestyle='--')
ax.set_title('Plano Factorial ACM')
ax.set_xlabel('Dimensión 1')
ax.set_ylabel('Dimensión 2')
ax.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
